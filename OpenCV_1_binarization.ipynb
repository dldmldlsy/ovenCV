```python
import cv2
from google.colab.patches import cv2_imshow 
```

```python
img_source1=cv2.imread('img_source1.jpg')
cv2_imshow(img_source1)
```

```python
img1_gray=cv2.cvtColor(img_source1,cv2.COLOR_RGB2GRAY)
cv2_imshow(img1_gray)
```

```python
thres, img1_result=cv2.threshold(img1_gray,150,255,cv2.THRESH_BINARY)
thres, img1_result_inv=cv2.threshold(img1_gray,150,255,cv2.THRESH_BINARY_INV)
cv2_imshow(img1_result_inv)
```

```python
#otsu algorithm
a, img_result_otsu=cv2.threshold(img1_gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
cv2_imshow(img_result_otsu)
```

```python
thres, img1_result=cv2.threshold(img1_gray,127,255,cv2.THRESH_BINARY)
cv2_imshow(img1_result)
```

```python
img_source2=cv2.imread('img_source2.png')
cv2_imshow(img_source2)
```

```python
img2_gray=cv2.cvtColor(img_source2,cv2.COLOR_RGB2GRAY)
cv2_imshow(img2_gray)
```

```python
thres, img2_result=cv2.threshold(img2_gray,150,255,cv2.THRESH_BINARY)
thres, img2_result_inv=cv2.threshold(img2_gray,150,255,cv2.THRESH_BINARY_INV)
cv2_imshow(img2_result_inv)
```

```python
thres, img2_result=cv2.threshold(img2_gray,150,255,cv2.THRESH_BINARY)
cv2_imshow(img2_result)
```

```python
img2_result_adapt=cv2.adaptiveThreshold(img2_gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,21,5)
cv2_imshow(img2_result_adapt)
```

```python
img_blur=cv2.GaussianBlur(img2_gray,(5,5),0)
img2_result_blur_adapt=cv2.adaptiveThreshold(img_blur,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,21,5)
cv2_imshow(img2_result_blur_adapt)
```
